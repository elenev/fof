```python
import pandas as pd
```

Read data:

```python
df = pd.read_csv('FRB_Z1_test.csv')
dfdict = pd.read_csv('FRB_Z1_dict.csv')
```

Merge dictionary with data and process dates:

```python
dfcombo = dfdict.merge(df,on='series_code')
dfcombo['time_period'] = pd.to_datetime(dfcombo['time_period'], format='%Y-%m-%d')
dfcombo.tail()
```

If neeeded, reshape the data long to wide:

```python
dfwide = dfcombo.pivot(index='time_period', columns='series_code', values='obs_value')
dfwide.tail()
```
